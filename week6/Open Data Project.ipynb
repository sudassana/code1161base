Shows the types of the dataset ^^
But I need Year_of_Release to be int64 rather than float64

```python
vgsales_data['Year_of_Release'] = vgsales_data['Year_of_Release'].fillna(0).astype(np.int64)
vgsales_data.head()
```

For example, I want to pick up the information of row seventeenth, I should type this

```python
row_seventeenth = vgsales_data.iloc[17]
row_seventeenth
```

```python
vgsales_data["Platform"].value_counts().plot(kind="bar")

plt.xlabel('Platform', fontsize = 18)
plt.ylabel('Video Games', fontsize = 18)
plt.title('Video Games Published on Each Platform', fontsize = 20)
```

```python
vgsales_data["Genre"].value_counts().plot(kind= "bar")

plt.title('Video Games Published on Each Genre', fontsize = 20)
plt.xlabel('Genre', fontsize = 18)
plt.ylabel('Video Games', fontsize = 18)
```

What about the score about this specific game? Put Critic Score inside the bracket.

```python
gen_tot=vgsales_data[['Genre','Global_Sales']].groupby('Genre').sum()
gen_tot.plot(kind='bar')

plt.title('Global Sales based on Genre', fontsize=25)
plt.xlabel('Genre', fontsize=18)
plt.ylabel('Sales', fontsize=18)
plt.show()
```

```python
vgsales_data["Rating"].value_counts().plot(kind= "bar")

plt.title('Video Games Published on Rating', fontsize = 25)
plt.xlabel('Rating', fontsize = 18)
plt.ylabel('Video Games', fontsize = 18)
```

That is a lot of data in the histogram, now I will limit the score greater than 70

```python
year_sales=vgsales_data[["Year_of_Release","Global_Sales"]].groupby('Year_of_Release').sum()
year_sales.plot(kind='bar')

plt.title('Global Sales per Year', fontsize = 20)
plt.xlabel('Year', fontsize = 18)
plt.ylabel('Sales', fontsize = 18)
```

But before, what if the Critic_Score was in plot diagram?

```python
reg_sales=vgsales_data[["Year_of_Release","NA_Sales", "EU_Sales", "JP_Sales"]].groupby('Year_of_Release').sum()
reg_sales.plot(kind='bar')

plt.title('Region Sales Comparison per Year', fontsize = 20)
plt.xlabel('Year', fontsize = 18)
plt.ylabel('Sales', fontsize = 18)
```

Let's see what is the histogram looks like for GOTY worthy games (critic score > 90)

```python
pub_sales=vgsales_data[["Publisher","Global_Sales"]].groupby('Publisher').sum()
pub_sales[pub_sales > 1000].plot(kind="bar")
```
