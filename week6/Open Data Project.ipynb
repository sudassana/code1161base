```python
vgsales_data.dtypes
#What ^^ does is to check the type of input on each columns, some types cannot be converted into a diagram.
```

```python
vgsales_data['Year_of_Release'] = vgsales_data['Year_of_Release'].fillna(0).astype(np.int64)
vgsales_data.head() 
#Open a preview of the table
```

```python
row_seventeenth = vgsales_data.iloc[17]
row_seventeenth
```

```python
vgsales_data["Platform"].value_counts().plot(kind="bar")

plt.xlabel('Platform', fontsize = 18)
plt.ylabel('Video Games', fontsize = 18)
plt.title('Video Games Published on Each Platform', fontsize = 20)
```

```python
comp1=vgsales_data[["Platform","Global_Sales"]].groupby('Platform').sum()
comp1.plot(kind='bar', facecolor= 'cyan')

plt.title('Video Games Sales on Each Platform', fontsize = 22)
plt.xlabel('Platform', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
```

```python
vgsales_data["Platform"].value_counts().plot(kind="bar")

plt.xlabel('Platform', fontsize = 18)
plt.ylabel('Video Games', fontsize = 18)
plt.title('Video Games Published on Each Platform', fontsize = 20)
```

```python
vgsales_data["Genre"].value_counts().plot(kind= "bar", facecolor='red')

plt.title('Video Games Published on Each Genre', fontsize = 22)
plt.xlabel('Genre', fontsize = 20)
plt.ylabel('Video Games', fontsize = 20)
```

```python
gen_tot=vgsales_data[['Genre','Global_Sales']].groupby('Genre').sum()
gen_tot.plot(kind='bar', facecolor='red')

plt.title('Global Sales based on Genre', fontsize=22)
plt.xlabel('Genre', fontsize=20)
plt.ylabel('Sales', fontsize=20)
```

```python
vgsales_data["Rating"].value_counts().plot(kind= "bar", facecolor='yellow')

plt.title('Video Games Published on Rating', fontsize = 22)
plt.xlabel('Rating', fontsize = 20)
plt.ylabel('Video Games', fontsize = 20)
```

```python
year_sales=vgsales_data[["Year_of_Release","Global_Sales"]].groupby('Year_of_Release').sum()
year_sales.plot(kind='bar', facecolor='green')

plt.title('Global Sales per Year', fontsize = 22)
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
```

```python
reg_sales=vgsales_data[["Year_of_Release","NA_Sales", "EU_Sales", "JP_Sales"]].groupby('Year_of_Release').sum()
reg_sales.plot(kind='bar')

plt.title('Region Sales Comparison per Year', fontsize = 22)
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
```

```python
plt.hist(vgsales_data["Critic_Score"], range=(0, 100))
plt.title('Video Games Critic Score', fontsize = 22)
plt.xlabel('Critic Score', fontsize = 20)
plt.ylabel('Video Games', fontsize = 20)
plt.grid(True)
plt.show()
```

```python
vgsales_data["Critic_Score"][vgsales_data["Critic_Score"] > 80].hist()

plt.title('Critic Score > 80', fontsize = 22)
plt.xlabel('Critic Scores', fontsize = 20)
plt.ylabel('Video Games', fontsize = 20)
```

```python
pub_sales=vgsales_data[["Publisher","Global_Sales"]].groupby('Publisher').sum()
pub_sales
```
